```python
import tarfile,sys,pandas as pd
from io import StringIO
```

```python
test_df_neg = pd.DataFrame()
test_df_pos = pd.DataFrame()
train_df_neg = pd.DataFrame()
train_df_pos = pd.DataFrame()

tar = tarfile.open("aclImdb_v1.tar.gz",'r:gz')

for member in tar.getmembers():
    
    #Creating negative test dataframe
    if "aclImdb/test/neg/" in member.name:
        #print(member.name)
        f=tar.extractfile(member)
        content = f.read()

        #Converting byte to string using UTF-8 format
        s=str(content,'utf-8')
        data=list(StringIO(s))
        
        #Appending a dataframe
        test_df_neg=test_df_neg.append(data)
        test_df_neg['label'] = '0'
         #Creating positive test dataframe    
    elif "aclImdb/test/pos/" in member.name:
        #print(member.name)
        f=tar.extractfile(member)
        content = f.read()

        #Converting byte to string using UTF-8 format
        s=str(content,'utf-8')
        data=list(StringIO(s))
        
        #Appending a dataframe
        test_df_pos=test_df_pos.append(data)
        test_df_pos['label'] = '1'
        
    #Creating negative train dataframe
    elif "aclImdb/train/neg/" in member.name:
        #print(member.name)
        f=tar.extractfile(member)
        content = f.read()

        #Converting byte to string using UTF-8 format
        s=str(content,'utf-8')
        data=list(StringIO(s))
         #Appending a dataframe
        train_df_neg=train_df_neg.append(data)
        train_df_neg['label'] = '0'
        
    #Creating positive train dataframe
    elif "aclImdb/train/pos/" in member.name:
        #print(member.name)
        f=tar.extractfile(member)
        content = f.read()

        #Converting byte to string using UTF-8 format
        s=str(content,'utf-8')
        data=list(StringIO(s))
        
        #Appending a dataframe
        train_df_pos=train_df_pos.append(data)
        train_df_pos['label'] = '1'
        
tar.close()
```

```python
test_df=pd.DataFrame()
test_df = test_df_neg
test_df = test_df.append(test_df_pos)
```

```python
test_df = test_df.rename(columns = {0:'review'}).reset_index(drop=True)
test_df
```

```python
test_df.to_csv("test1.csv",sep=',')
```
